### Read in mapping file from geoID to census tract

```python
url1 = 'http://www2.census.gov/geo/maps/dc10map/tract/st36_ny/c36061_new_york/DC10CT_C36061_CT2MS.txt'
response1 = urllib2.urlopen(url1)

url2 = 'http://www2.census.gov/geo/maps/dc10map/tract/st36_ny/c36047_kings/DC10CT_C36047_CT2MS.txt'
response2 = urllib2.urlopen(url2)

url3 = 'http://www2.census.gov/geo/maps/dc10map/tract/st36_ny/c36081_queens/DC10CT_C36081_CT2MS.txt'
response3 = urllib2.urlopen(url3)

url4 = 'http://www2.census.gov/geo/maps/dc10map/tract/st36_ny/c36085_richmond/DC10CT_C36085_CT2MS.txt'
response4 = urllib2.urlopen(url4)

url5 = 'http://www2.census.gov/geo/maps/dc10map/tract/st36_ny/c36005_bronx/DC10CT_C36005_CT2MS.txt'
response5 = urllib2.urlopen(url5)
```

```python
map1=pd.read_csv(response1, sep=';')
map2=pd.read_csv(response2, sep=';')
map3=pd.read_csv(response3, sep=';')
map4=pd.read_csv(response4, sep=';')
map5=pd.read_csv(response5, sep=';')
```

#### Concating five boroughs

```python
frames = [map1, map2, map3, map4, map5]
mapnyc = pd.concat(frames)
```

#### combine NAME and SHEETS column to sync with census tract column

```python
mapnyc.columns = ['TYPE','GEO.id2','Census','SHEETS']
mapnyc
```

```python
acs_mapped = pd.merge(acs, mapnyc, on='GEO.id2', how='inner')
```

```python
del acs_mapped['TYPE']
del acs_mapped['SHEETS']
```

## census tracts to pums data

```python
lis = []
for value in acs['GEO.id2']:
    lis.append(int(value)-36005000000)

acs['census']=lis
```
